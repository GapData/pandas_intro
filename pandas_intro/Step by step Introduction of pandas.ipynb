```python
# 修改一下第一欄的名字
col1 = '人口密度'
TWdata.columns = [col1, TWdata.columns[1], TWdata.columns[2]]
TWdata
```

### pandas.DataFrame 是你在 Python 中處理「表格」資料最好的幫手了！

```python
# 讓我們快速驗證人口密度是怎麼算的吧！ 人口密度 ＝ 人口/ 地區面積
TWdata['population density'] = TWdata[TWdata.columns[1]] / TWdata[TWdata.columns[2]] # 夠直覺吧！
TWdata
```

```python
# 先對我們計算的人口密度取到小數點第二位
TWdata['population density'] = TWdata['population density'].apply(lambda x: round(x, 2)) 
# 在 Series 和 DataFrame 裡面，我們是用 apply 而不是 map 喔
TWdata
```

```python
# 來抓出哪些資料是我輸入錯誤的吧 ><''
TWdata[(TWdata[TWdata.columns[0]] == TWdata['population density']) == False]
```
